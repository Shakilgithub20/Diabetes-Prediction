```python
# drop columns number 9
train_data_x = dataset.drop(columns = 'Outcome', axis = 1)
train_data_y = dataset['Outcome']
```

```python
# split into 80% for train and 20% for test
# random number generator 10
train_Data, test_Data, train_Out, test_Out = train_test_split(train_data_x,train_data_y,test_size=0.2, random_state=10)
print(train_Out)
```

```python
# create model
# using model keras
# using activations functions sigmoid
data_model = keras.Sequential(
    [
      #  keras.layers.Flatten()
      keras.layers.Dense(units=100, input_dim=8, name='hidden_layer_1'),
      keras.layers.Dense(units=90, name='hidden_layer_2'),
      keras.layers.Dense(units=80, name='hidden_layer_3'),
      keras.layers.Dense(units=70, name='hidden_layer_4'),
      keras.layers.Dense(units=60, name='hidden_layer_5'),
      keras.layers.Dense(units=50, name='hidden_layer_6'),
      keras.layers.Dense(units=40, name='hidden_layer_7'),
      keras.layers.Dense(units=30, name='hidden_layer_8'),
      keras.layers.Dense(units=1, activation=tf.keras.activations.sigmoid, name='output_layer')
    ]
)

print(data_model.summary())
```

```python
# Compile model
# using opimizer Adam
# using loss binary_crossentropy
data_model.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.binary_crossentropy, metrics=['accuracy'])
```

```python
data_model.fit(train_Data, train_Out, validation_data=(test_Data, test_Out), epochs=1000, verbose=1)
```

```python
# Final Prediction
print(data_model.predict(test_Data))
```
